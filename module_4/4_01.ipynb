### Figures y Subplots
- Varias formas de añadir varios gráficos a una misma figura.
  * Con plt.figure() y fig.add_subplot
  * Con plt.subplots y usando los axes. Es la opción recomendada.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
fig, axs = plt.subplots(2, 2)
```

```python
axs
```

```python
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2)
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax2.plot(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax4.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
#### ajustar el tamaño de la figura
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 10))
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax4.plot(np.random.randn(50).cumsum(), 'k--')
```

#### Añadiendo legends.

```python
fig, ax = plt.subplots(1, figsize=(10,7))
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend()
```

### Anotaciones  y Dibujos en un subplot

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)


fig, ax = plt.subplots(1, figsize=(10,7))
ax.plot(x, y, label='one')
ax.annotate('En x= $\pi$', 
            xy=(np.pi, 0.01),
            xytext=(np.pi, 0.3),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
            horizontalalignment='left',
            verticalalignment='top')
```

- Podemos poner figuras.

```python
fig, ax = plt.subplots(figsize=(12, 6))
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

- Para seguir utilizaremos una base de datos de propinas en resturantes:

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
tips.shape
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
# Not many 1- and 6-person parties
party_counts = party_counts.loc[:, 2:5]
party_counts
```

```python
# Normalize to sum to 1
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
# boxplot
tips[['total_bill', 'tip']].boxplot()
```

- Podemos pasar un ax para modificar y hacer subplots.

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(13,7))
party_pcts.plot.bar(ax=ax1)
ax1.set_title('Bars')
tips[['total_bill', 'tip']].boxplot(ax=ax2)
ax2.set_title('Box')
```

### Ploting de datos financieros

- Adquirimos datos financieros: apple stock.

```python
import urllib.request
import certifi
import json
```

```python
url = 'https://api.iextrading.com/1.0/stock/aapl/chart/5y'
contents = urllib.request.urlopen(url, cafile=certifi.where()).read()
```

```python
datos_apple = pd.read_json(contents)
datos_apple.index = datos_apple.date
datos_apple = datos_apple[['open', 'high', 'low', 'close', 'volume']]
datos_apple = datos_apple['2018':]
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
datos_apple.close.plot()
```

- Para obtener gráficos de velas tenemos varias posibilidades:
 * usando la librería mpl_finance.
 * col la librería plotly.


```python
!pip install mpl_finance
```

```python
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
datos_apple = datos_apple['2019']
datos_ohlc = datos_apple.iloc[:, :4]
datos_ohlc['date'] = pd.to_datetime(datos_ohlc.index)
datos_ohlc["date"] = datos_ohlc["date"].apply(mdates.date2num)
datos_ohlc = datos_ohlc[['date', 'open', 'high', 'low', 'close']]
datos_ohlc = datos_ohlc.values
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
_ = candlestick_ohlc(ax, datos_ohlc, width=.5, colorup='green', colordown='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd
from datetime import datetime
```

```python
url = 'https://api.iextrading.com/1.0/stock/aapl/chart/5y'
contents = urllib.request.urlopen(url, cafile=certifi.where()).read()

datos_apple = pd.read_json(contents)
datos_apple.index = datos_apple.date
datos_apple = datos_apple[['open', 'high', 'low', 'close', 'volume']]

trace = go.Candlestick(x=datos_apple.index,
                       open=datos_apple['open'],
                       high=datos_apple['high'],
                       low=datos_apple['low'],
                       close=datos_apple['close'])
data = [trace]
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=False)
iplot(data)
```
